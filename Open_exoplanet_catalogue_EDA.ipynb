## The basics

```python
# First we import some useful packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
#plt.rcParams['figure.figsize'] = (15, 15)
```

```python
# Loading the data.
exoplanets = pd.read_csv('oec.csv')
```

Let's see what the data looks like.

```python
exoplanets
```

The original csv is comma separated but Pandas (and Jupyter) displays it in a pretty tabular form. The index on the left is added by Panda. So we have 3584 rows x 25 rows, which corresponds to our 3584 exoplanets and the 25 attributes.

By default Pandas (thankfully) doesn't diplays all of the data, but it provides an eay way to get the name of the columns. 

```python
exoplanets.columns
```

An another useful command is the .info() function.

```python
exoplanets.info()
```

As you can see, it gives you a summary of your dataset, with the number of entries, columns, description of each column, the types in the dataset, plus the memory it uses to hold the data.

If you look at the csv file, you can find series of comma, with no values inbetween (Pandas replaces thoses blanks by NaN) suggesting we are missing some data. We can now confirm that and put a number on it with exoplanets.info(). 
"PlanetIdentifier 3584 non-null object" tells us that we have 3584 entries of type object in that column, so it's not missing anything, whereas "PlanetaryMassJpt  1313 non-null float64" tells us we are missing quite a bit.

Actually, if you look at all the columns, only 3 of thems are full. AgeGyr is basically empty with only 2 entries.
Why? I don't know. Maybe the dataset is incomplete or if the actual values are just not there. In any case we'll have to deal with the missing values moving forward.

## Let's start plotting some stuff.

First let's see an how many planets were discovered each year. 

Pandas provides an histogram through DataFrame.hist().It needs the dataframe, the column and the number of bins you want to use (optionnal).

```python
pd.DataFrame.hist(exoplanets, 'DiscoveryYear', bins=236)
```

The graph starts before 1800, which is not supposed to happen, the first exoplanets being discovered in the 1990s.
If you look at the stats about DiscoveryYear, you can see the min is 1781. 

```python
exoplanets[exoplanets['DiscoveryYear'] == 1781]
```

As you can see they added Uranus, apparently discovered in 1781. 
Let's see if other planets from the Solar System are present.

```python
exoplanets[exoplanets['ListsPlanetIsOn'] == 'Solar System']
```

Now that we know that the entire Solar System is included in the dataset, we'll have to filter them out.

We only want planets discovered from 1990 going forward. One way to do it is to use the range option from pd.DataFrame.hist.

```python
pd.DataFrame.hist(exoplanets, 'DiscoveryYear', bins=27, range=(1990,2017))
```
